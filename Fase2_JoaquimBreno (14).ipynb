Implemente uma Lista em Python utilizando os métodos: \\
* create_list() - Criar uma lista vazia. \\
* check_empty() - Checar se uma lista está vazia. \\
* append() - Adicionar um elemento na lista. \\
* pop() - Remover o elemento mais antigo da lista. \\

\\
Referências:

[1] https://docs.python.org/3/tutorial/datastructures.html

Você deve adicionar na lista de processos os seguintes elementos (strings) abaixo: \\

    1 - "Ler request do client"
    2 - "Processar request"
    3 - "Chamar função específica"
    4 - "Formatar resultados"
    5 - "Retornar para o client"

Depois de adicionado cada elemento realize a operação de pop() e dê um print do processo atual.

```python
# Definindo uma tabela com todos os processos
processos = {
    'strings': ["Ler request do client", "Processar request", "Chamar função específica", "Formatar resultados", "Retornar para o client"]
}
```

```python


# Etapa 1) Criar um método main:
def main():
  list = create_list()
  checker = check_empty(list)
  list = append(list, processos) 
  list = pop(list)
  print(checker)
  print('Processo mais recente:  ' + list[-1])

# Etapa 2) Criar um método para criar uma empty list:
def create_list():
  list = []
  return list

# Etapa 3) Criar um método para verificar se listas estão vazias:
def check_empty(list):
  if not list:  #Condição para caso a lis
    return "Essa lista estava vazia"
  else:
    return "Havia items nessa lista"

# Etapa 4) Iniciar a adição de items-instruções dentro da lista:
def append(list,processos):
  c = 0
  for i in processos['strings']: # Convida a tabela de processos e navega entre as strings da primeira à última
    c += 1
    list.append(str(c) + " - " + i) # Concatena a posição e a string, adicionando-as na lista
  return list

# Etapa 5) Remover elemento mais antigo da lista:
def pop(list):
  list.pop(0)
  return list

# Executa todas as etapas
if __name__ == '__main__':
    main()
```

```python
def definetitulo(x):
  titulo = x["Titles"]
  if titulo in ['Capt', 'Col', 'Don', 'Jonkheer', 'Major', 'Rev', 'Sir']:
    return 'Mr'
  elif titulo in ['the Countess', 'Mme', 'Lady']:
    return 'Mrs'
  elif titulo in ['Mlle', 'Ms']:
    return 'Miss'
  elif titulo =='Dr':
    if x['Sex']=='male':
      return 'Mr'
    else:
      return 'Mrs'
  else:
    return titulo

dataset["Titles"]= dataset.apply(definetitulo, axis=1)
#dataset["Titles"]

titulos = set(dataset.Titles)
print(titulos)
```

_Sua resposta aqui_


```python
# O uso do natural language tool kit facilita o processo de tokenizar palavras
# assim, pode-se usar a função embutida na biblioteca "word_tokenize" e tokenizar palavras
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
sms["Token"] = sms["TextClean"].map(lambda x: word_tokenize(x)) # mapeia todas as palavras do texto limpo e tokeniza
sms["Token"]
```

3.2.3 - Remover stop words

_Sua resposta aqui_


```python
# O uso do natural language tool kit facilita o processo de busca de stopwords
# com um banco de stopwords extensos, a ferramenta oferece a opção de navegar por palavras
# tanto em português quanto em inglês
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) # Escolhi em inglês pois os sms estão em inglês
def tirar_stopwords(sentence):
    clean_tokens = [w for w in sentence if not w in stop_words] # o loop de navegação entre as palavras do dataframe sms permite apenas palavras não disponiveis nas stopwords
    return clean_tokens
    
sms['OutStop'] = sms['Token'].apply(tirar_stopwords)
sms['OutStop']
```

3.2.4 - Lematização

```python
# O uso do natural language tool kit facilita o processo de lematização
# com o wordnetlemmatizer é possível identificar radicais de palavras pelo
# contexto a qual estão aplicadas
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet = WordNetLemmatizer()
def lemmatizer(sentence):
    clean_tokens = [wordnet.lemmatize(w) for w in sentence]   # O mesmo processo verificado nas stopwords se repete, agora com a lemmatização de cada palavra
    return clean_tokens
    

sms['Lemma'] = sms['OutStop'].apply(lemmatizer)
sms['Lemma']
```

## Faça uma base de dados com as estatísticas dos posts do Instagram da Tail, converta em csv, limpe dados faltantes e faça análise exploratória.

```python
# Bibliotecas usadas para fazer o scrapping da página 
!pip install kora -q
from kora.selenium import wd

import pandas as pd
import time
```

```python
# Tentei executar o selenium na virtualenv do Google Collab, contudo foram várias tentativas por água abaixo
# Aparentemente erros de log como " BLuetooth adapter " não são ignorados pelo Collab 
# Mas enfim, deixei o código exposto e irei apenas importar o dataset "tailscrp"
class InstagramBot:
    

    def __init__(self): # Chamo a função web driver "wd"
        self.driver = wd
 
    def feedtail(self):
        driver = self.driver
        driver.get("https://www.instagram.com/tailufpb/") #Abre a página da tail
        time.sleep(3)
        
        for i in range(0,3): # Dá scroll na página em busca de publicações
            driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
            time.sleep(2)

        name = driver.find_elements_by_class_name('v1Nh3') # Recebe as publicações por class name
        hrefs = [elem.find_element_by_css_selector('a').get_attribute('href') for elem in name] # Guarda os endereços das publicações

        index = [] # Abertura de duas listas com index e colunas do dataframe
        for c in range(0, len(hrefs)):
            index.append(c) # Cria index em relação aos hrefs coletados
        columns = ['Data', 'Likes', 'Comentários', 'Descrição'] 
        tailscrp = pd.DataFrame(index=index, columns=columns)
        tailscrp = tailscrp.fillna(0) # Valores nulos no dataset são alterados por 0
        c = -1
        

        for href in hrefs: # Inicia o processo de coleta de dados em cada postagem
            c += 1
            driver.get(href)
            tailscrp.loc[c,'Data'] = driver.find_element_by_class_name('_1o9PC').get_attribute('datetime') # Recebe a data da postagem
            try: # Recebe o valor dos likes
                tailscrp.loc[c,'Likes'] = driver.find_element_by_class_name('zV_Nj').find_element_by_css_selector('span').get_attribute('innerHTML')
            except: # Caso a postagem seja um vídeo a classe muda, portanto deve-se procurar por essa classe e receber os valores dos likes
                tailscrp.loc[c,'Likes'] = driver.find_element_by_class_name('vcOH2').find_element_by_css_selector('span').get_attribute('innerHTML')
            lists = driver.find_elements_by_class_name('XQXOT')
            for list in lists: # Aqui vou varrer pelos comentários e descrição dando um scroll
                driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', list)
                time.sleep(3)
            tailscrp.loc[c, 'Comentários'] = len(driver.find_elements_by_class_name('C7I1f')) # Recebe os comentários
            tailscrp.loc[c, 'Descrição'] = driver.find_element_by_class_name('C4VMK').find_element_by_css_selector('span').get_attribute('innerHTML') # Recebe a descrição
        
        tailscrp.to_csv(r'C:\Users\JoaquimBreno\Documents\Programming\Python\Scrapper Tail\tailscrp.csv', index = False, header=True, na_rep="None") # Converte em csv
        print(tailscrp)


jokaBot = InstagramBot() # Inicia a classe
jokaBot.feedtail() # Executa o métdo feedtail
```

```python
tailscrp = pd.read_csv('https://raw.githubusercontent.com/JoaquimBreno/datasets/master/tailscrp.csv', sep = ',')
```

```python
tailscrp
```

```python
# Podemos ver que houveram 28 publicações até o momento do dia 26/03/2021
# A data está mal formatada, adjunta com as horas , minutos e segundos da postagem
# Irei formatá-la a seguir
```

```python
tailscrp['Hora'] = tailscrp.Data.apply(lambda x: x[11:19])
```

```python
tailscrp['Data'] = tailscrp.Data.str[0:10:1]
```

```python
tailscrp = tailscrp[['Data','Hora','Likes','Comentários','Descrição']]
tailscrp
```

```python
# Agora irei limpar um pouco a descrição de cada postagem
```

```python
import re
tailscrp["Descrição"] = tailscrp[["Descrição"]]\
.apply(lambda x: x.astype(str).str.lower())\
.apply(lambda x: x.astype(str).str.normalize("NFKD").str.encode("ascii", errors="ignore").str.decode("utf-8")) 
for x in range(len(tailscrp["Descrição"])):
  tailscrp.loc[x, "Descrição"] = re.sub('\W+',' ', tailscrp.loc[x, "Descrição"])
tailscrp["Descrição"]
```

```python
# É necessário, também, tirar algumas tags que vieram junto com a formatação html 
# Irei transformar as palavras em tokens:
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
tailscrp['Descrição'] = tailscrp['Descrição'].map(lambda x: word_tokenize(x)) 
tailscrp['Descrição']
```

```python
# Criei uma lista com as palavras indesejadas
```

```python
tagnames = ['br','xil3i','tabindex','href','a','class','tags', 'ai', '0']
```

```python
# E naveguei em todos as palavras dos tokes em busca de palavras que não têm as tag names
```

```python
def limpartags(tokens):
    clean = [palavras for palavras in tokens if not palavras in tagnames] 
    return clean
    
tailscrp['Descrição'] = tailscrp['Descrição'].apply(limpartags)
tailscrp['Descrição']
```

```python
# Vou criar uma coluna com o número de palavras de cada descrição que posteriormente pode servir pra alguma coisa
```

```python
tailscrp['numPalavras'] = 0
count = 0
row = -1
for tokens in tailscrp['Descrição']:
  row+=1
  for palavras in tokens:
    count+=1
  tailscrp.loc[row, 'numPalavras'] = count
```

```python
# Por fim, untokenizei tudo
```

```python
from nltk.tokenize.treebank import TreebankWordDetokenizer
tailscrp['DescriçãoTok'] = tailscrp['Descrição'].map(lambda x: TreebankWordDetokenizer().detokenize(x)) 
```
